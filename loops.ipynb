Break and Continue

```python
fruits = ['apple','mango','banana','grapes','tomato','papaya','orange','kiwi']
for x in fruits:
    if x == 'tomato':
        break
    print(x)
```

```python
fruits = ['apple','mango','banana','grapes','tomato','orange','kiwi']
for x in fruits:
    if x == 'tomato':
        continue
    print(x)
```

```python
num = 7
if num < 2:
    print("Not Prime number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime number")
            break
    else:
        print("Prime number")
```

```python
n = 4
if n < 2:
    print("Not prime")
else:
    for i in range(2, n):
        if n % i == 0:
            print("not prime")
            break
    else:
        print("prime")        
```
