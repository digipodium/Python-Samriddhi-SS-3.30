```python
import math
```

```python
print(math.sqrt(16))
print(math.factorial(5))
print(math.gcd(12, 15))
print(math.lcm(12, 15))
print(math.pi)
print(math.log(100, 10))
print(math.pow(2, 3))
```

```python
from random import random, randint, choice, shuffle
```

```python
print("number between 0 and 1:", random())
print("random integer between 10 and 20:", randint(10, 20))
print("random choice from a list:", choice(['apple', 'banana', 'cherry']))
name = ['Alice', 'Bob', 'Charlie', 'David']
shuffle(name)
print("shuffled list:", name)
```

```python
from datetime import datetime as dt #alias for datetime module  
```

```python
print("Current date and time:", dt.now())
print("Current Date:", dt.now().date())
print("Current Time:", dt.now().time())
```

```python
print("Current Year:", dt.now().year)
print("Current Month:", dt.now().month)
print("Current Day:", dt.now().day)
print("Current Hour:", dt.now().hour)
print("Current Minute:", dt.now().minute)
print("Current Second:", dt.now().second)
```

```python
print("Print Current Date ", dt.now().strftime("%d / %m / %Y"))
print("Print Current Time ", dt.now().strftime("%H:%M:%S"))
print("Print Current Month name :", dt.now().strftime("%A"))
```
