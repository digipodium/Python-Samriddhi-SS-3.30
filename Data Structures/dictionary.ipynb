```python
student = {
    "name" : "Alice",
    "age" : 20,
    'course' : 'Computer Science',
    'school' : 'XYZ University'
}
print(student)
```

```python
temp = {
    'lucknow' : [32, 45, 35],
    'chennai' : [30, 33, 31],
    'delhi' : [28, 34, 29]
}
print(temp)
```

```python
info = dict(name = "Bob", age = 20, course = 'Computer Science', school = 'ABC University')
print(info)
```

```python
info = dict(name = input("Enter a name: "), age = int(input("Enter age: ")), course = input("Enter course: "), school = input("Enter school: "))
print(info)
```

```python
#1st method to access values in a dictionary
info['name']
```

```python
#2nd method to access values in a dictionary
info.get('age')
```

```python
#to add a new key value pair 
info['grade'] = 'A'
print(info)
```

```python
#to delete a key value pair
info.pop('course')
print(info)
```

```python
#to update an existing key value pair
info['age'] = 21
print(info)
```

```python
print(info.keys())
print(info.values())
print(info.items())
```

```python
for v in info.values():
    print(v)
```

```python
for k,v in info.items():
    print(f"{k} : {v}")
```
