```python
apps = ['Facebook', 'Instagram', 'Twitter', 'LinkedIn','Snapchat','swiggy','zomato']
print(apps)
```

INDEXING & SLICING

```python
print(apps[2])
print(apps[4])
```

```python
print(apps[-3])
```

```python
print(apps[1:6])
```

```python
print(apps[1:8:2])
```

```python
print(apps[:]) #it will give a complete list
print(apps[:8]) #it will give a list from start to 8th element
print(apps[3:]) #it will give a list from 3rd element to end
print(apps[::2]) #it will give a list from start to end with step 2
print(apps[-4:-1]) #it will give a list from -4th element to -1th element
print(apps[-4:]) #it will give a list from -4th element to end
```

LIST METHODS
- append() - to add item at the end of the list
- insert() - to add item at any specific position in the list
- pop() - to delete item using its index position
- remove() - to delete items using item name
- copy() - to make a duplicate of the list
- sort() - to sort the list in a particular order
- extend() - to combine two list into a single list
- clear() - to delete all the items from the list but not the list

```python
apps.append("Netflix")
print(apps)
```

```python
apps.insert(3, "Ola")
print(apps)
```

```python
apps.pop(3)
print(apps)
```

```python
apps.remove("Twitter")
print(apps)
```

```python
copy = apps.copy()
print(copy)
```

```python
copy.clear()
print(copy)
```

```python
apps.sort()
print(apps)
```

```python
apps2 = [2,3,4,5]
apps.extend(apps2)
print(apps)
```

```python
apps2.extend(apps)
print(apps2)
```
