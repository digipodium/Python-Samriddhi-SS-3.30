- when you have >40% missing values in a column, you can simple drop the whole column untill n unless that column is very important for any type of analysis
- when you have <40% missing values then you can handle those missing values
- when you have 2 or 3 missing values in the whole dataset then we can simply remove those rows

there are two cases to handle missing values-
- numeric data column - we always replace those missing values with the average value of the column
- categorical data column - we always replace those missing values with the most frequent value of that column

```python
average_age = round(df['age'].mean())
print(average_age)
```

```python
df['age'] = df['age'].fillna(average_age)
df.isnull().sum()
```

```python
#delete rows with missing values in 'embarked' column
df = df.dropna(subset=['embarked'])
df.isnull().sum()
```
