```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.info()
```

```python
#Histogram - Distribution of Sepal Length
plt.hist(df['sepal_length'],color='skyblue',edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Line Plot - Petal length of first 50 samples
plt.plot(df['petal_length'][:50], color='orange')
plt.title('Petal Length of First 50 Samples')
plt.xlabel('Sample Index')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
#using groupby command in pandas to calculate mean values for each species  
mean_value= df.groupby('species').mea
```

```python
#bar chart - average feature values per species
mean_value= df.groupby('species').mean()
mean_value.plot(kind='bar', figsize=(5,5))
plt.title('Average Feature Values per Species')
plt.ylabel('Average Value (cm)')
plt.xticks(rotation=-20)
plt.show()
```

```python
# Creating a figure and axis object
fig, ax = plt.subplots(figsize=(6,6))
#setting the color of figure and axes background
fig.set_facecolor('yellow')
ax.set_facecolor('lightyellow')
x = [1,2,3,4,5,6,7,8,9]
y = [3,6,2,4,7,5,9,8,2]
ax.plot(x,y,
    color = 'red',
    linewidth=2,
    linestyle='--',
    marker='o',
    markerfacecolor='blue',
    markeredgecolor='black',
    markersize=8,
    markeredgewidth=2)
ax.set_title('Customized Line Chart', color='blue', fontsize=16)
ax.set_xlabel('X-axis', color='green', fontsize=14)
ax.set_ylabel('Y-axis', color='purple', fontsize=14)

#add text
ax.text(4.5,7.2, 'Sample Text', color='brown', fontsize=12)

#horizontal and vertical lines
ax.axhline(y=5, color='gray', alpha=0.2)
ax.axvline(x=5, color='gray', alpha=0.2)
```
