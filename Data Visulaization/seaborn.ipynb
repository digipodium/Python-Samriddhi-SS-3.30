```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
#age distribution
sns.histplot(df, x='age',bins=30, color='red', kde=True) #kde adds a smooth curve (kernel density estimate)
```

```python
import matplotlib.pyplot as plt
#age distribution
sns.histplot(df, x='age',bins=30, color='red', kde=True) #kde adds a smooth curve (kernel density estimate)
plt.xticks(range(0 , 81, 5))  # Set x-ticks from 0 to 80 with a step of 5
plt.show()
```

```python
#countplot - number of passengers by gender
sns.countplot(df, x='sex', palette='viridis')
plt.title('Number of Passengers by Gender')
plt.show()
```

```python
#count of passenger class by gender
sns.countplot(df, x='pclass', hue='sex')
plt.title('Number of Passengers by Class')
plt.show()
```

```python
#Scatter plot - Age vs fare(colored by gender)
sns.scatterplot(df, x='age', y='fare', hue='sex', palette='husl')
plt.title('Age vs Fare by Gender')
plt.show()
```
