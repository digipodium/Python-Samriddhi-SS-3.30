```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
```

```python
#total accidents by state
fig = px.bar(df, x='abbrev', y='total',
             title='Total Car Accidents by State',
             labels={'abbrev':'State', 'total':'Total Accidents'},
             template='plotly_dark')
fig.show()
```

```python
#top10 states by insurance premiums
# top10 = df.nlargest(10, 'ins_premium')
top10 = df.sort_values('ins_premium', ascending=False).head(10)
fig = px.bar(top10, x='abbrev', y='ins_premium',
             title='Top 10 States by Insurance Premiums',
             labels={'abbrev':'State', 'ins_premium':'Insurance Premiums'},
             template='plotly_dark',
             color='ins_premium',)
fig.show()
```

```python
#percentage of speeding accidents
fig = px.pie(df, values = 'speeding', names='abbrev',
             title='Percentage of Speeding Accidents by State',
             template='plotly_dark',
             color_discrete_sequence = px.colors.sequential.Purples_r)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
#Alcohol vs Speeding
fig = px.scatter(df, x='alcohol', y='speeding',
                 color='abbrev',
                 title='Alcohol vs Speeding Accidents by State',
                 labels={'alcohol':'Alcohol-related Accidents', 'speeding':'Speeding-related Accidents'},
                 template='plotly_dark',
                 size='total')
fig.update_layout(title={'x':0.5})
fig.show()
```

```python
#line chart of total accidents
fig = px.line(df, x='abbrev', y='total',
              title='Total Car Accidents by State',
              labels={'abbrev':'State', 'total':'Total Accidents'},
              template='plotly_dark',
              markers=True,
              line_shape='spline',
              color_discrete_sequence=px.colors.qualitative.Dark2,
              )
fig.show()
```
