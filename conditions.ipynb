```python
marks = 85
if marks >= 90:
    print("A+ Grade")
if marks <= 75:
    print("A Grade")
if marks <= 60:
    print("B Grade")
if marks < 60:
    print("Better luck next time")
```

```python
marks = 85
if marks >= 90:
    print("A+ Grade")
elif marks <= 75:
    print("A Grade")
elif marks <= 60:
    print("B Grade")
elif marks < 60:
    print("Better luck next time")
else:
    print("Invalid Marks")
```

```python
s1 = 5
s2 = 3
s3 = 5
if s1 == s2 == s3:
    print("Triangle is an equilateral triangle")
elif s1 == s2 or s2 == s3 or s1 == s3:
    print("Triangle is an isosceles triangle")
else:
    print("Triangle is a scalene triangle")
```

```python
username = input("Enter your username: ")
email = input("Enter your email: ")
password = input("Enter the password")
confirm_pass = input("Enter the password again")

if len(username)>5 and len(username)<=15:
    if '@' in email and len(email)<=20:
        if len(password)>5 and password == confirm_pass:
            print("User registration successful!")
        else:
            print("Password must be at least 6 characters long and match the confirmation.")
    else:
        print("Invalid email address.")
else:
    print("Username must be between 6 and 15 characters long.")
```
